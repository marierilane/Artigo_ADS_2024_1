```python
import matplotlib.pyplot as plt

# Dados corrigidos das métricas
workload_types = ['Leve', 'Moderada', 'Pesada']
cpu_usage = [6.5, 16.9, 40.1]
memory_usage = [65.2, 65.3, 67.3]
disk_usage = [94.2, 94.6, 94.2]
swap_usage = [0.6, 1.0, 1.2]
timestamps = [8, 15, 40]
file_counts = [20, 60, 80]
line_counts = [363, 754, 1548]
code_line_counts = [194, 408, 715]

# Função para criar gráficos com labels nos valores
def create_bar_chart_with_labels(metric, values, ylabel, title):
    plt.figure(figsize=(10, 6))
    bars = plt.bar(workload_types, values, color=['blue', 'orange', 'red'])
    plt.xlabel('Tipo de Carga de Trabalho')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(axis='y')

    for bar in bars:
        height = bar.get_height()
        plt.annotate(f'{height}%',
                     xy=(bar.get_x() + bar.get_width() / 2, height),
                     xytext=(0, 3),  # 3 points vertical offset
                     textcoords="offset points",
                     ha='center', va='bottom')

    plt.show()

# Gráficos para cada métrica com labels nos valores
create_bar_chart_with_labels('CPU Usage', cpu_usage, 'Uso da CPU (%)', 'Comparação de Uso da CPU')
create_bar_chart_with_labels('Memory Usage', memory_usage, 'Uso de Memória (%)', 'Comparação de Uso de Memória')
create_bar_chart_with_labels('Disk Usage', disk_usage, 'Uso de Disco (%)', 'Comparação de Uso de Disco')
create_bar_chart_with_labels('Swap Usage', swap_usage, 'Uso de Swap (%)', 'Comparação de Uso de Swap')
create_bar_chart_with_labels('Timestamps', timestamps, 'Tempo (s)', 'Comparação de Tempo de Execução')
create_bar_chart_with_labels('File Counts', file_counts, 'Quantidade de Arquivos', 'Comparação de Quantidade de Arquivos')
create_bar_chart_with_labels('Line Counts', line_counts, 'Quantidade de Linhas', 'Comparação de Quantidade de Linhas')
create_bar_chart_with_labels('Code Line Counts', code_line_counts, 'Quantidade de Linhas de Código', 'Comparação de Linhas de Código')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dados das estatísticas
data = {
    'Métrica': ['CPU', 'Memory', 'Disk', 'Interrupt', 'Network', 'Swap', 'Process_count'],
    'Leve Mean': [2.275, 64.525, 94.2, 770558058, 6317515210, 0.6, 292.25],
    'Leve Stdev': [2.8929512036442415, 0.47169905660283024, 0.0, 86186.66737958952, 29116.20033589548, 0.0, 1.5],
    'Moderada Mean': [8.0, 65.125, 94.2, 784642533.5, 6327378770.5, 1.0, 285],
    'Moderada Stdev': [5.1278449319529077, 0.14999999999999936, 0.0, 93869.80086090876, 33544.63524420361, 0.0, 0.0],
    'Pesada Mean': [18.95, 66.125, 94.2, 795581621.5, 6336929189.75, 1.2, 259],
    'Pesada Stdev': [17.4433756729740645, 0.14999999999999936, 0.0, 70038.41685579898, 448.1988955809686, 0.0, 0.0]
}

# Criar dataframe
df = pd.DataFrame(data)

def create_bar_chart_for_metric(df, metric, ylabel, title):
    plt.figure(figsize=(12, 6))
    bar_width = 0.2
    bar_positions = [0, 1, 2]

    leve_value = df.loc[df['Métrica'] == metric, 'Leve Mean'].values[0]
    moderada_value = df.loc[df['Métrica'] == metric, 'Moderada Mean'].values[0]
    pesada_value = df.loc[df['Métrica'] == metric, 'Pesada Mean'].values[0]

    plt.bar([p - bar_width for p in bar_positions], [leve_value], width=bar_width, label='Leve', color='blue')
    plt.bar(bar_positions, [moderada_value], width=bar_width, label='Moderada', color='orange')
    plt.bar([p + bar_width for p in bar_positions], [pesada_value], width=bar_width, label='Pesada', color='red')

    plt.xlabel('Tipo de Carga de Trabalho')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks(bar_positions, ['Leve', 'Moderada', 'Pesada'])
    plt.legend()
    plt.grid(True)
    plt.show()

# Criar gráficos métrica a métrica
metrics_to_plot = ['Mean', 'Stdev']
for metric in df['Métrica']:
    for value_type in metrics_to_plot:
        create_bar_chart_for_metric(df, metric, f'{value_type} ({metric})', f'Comparação das {value_type} da {metric} por Tipo de Carga de Trabalho')

# Tabela de dados
table_df = pd.DataFrame(data)

# Mostrar tabela
print(table_df)

# Gráfico completo comparando todas as métricas para diferentes tipos de carga de trabalho
plt.figure(figsize=(14, 8))

for column in ['Leve Mean', 'Moderada Mean', 'Pesada Mean']:
    plt.plot(df['Métrica'], df[column], marker='o', label=column)
    for i, value in enumerate(df[column]):
        plt.annotate(f'{value:.2f}',
                     xy=(df['Métrica'][i], value),
                     xytext=(5, 5),
                     textcoords='offset points')

for column in ['Leve Stdev', 'Moderada Stdev', 'Pesada Stdev']:
    plt.plot(df['Métrica'], df[column], marker='x', linestyle='--', label=column)
    for i, value in enumerate(df[column]):
        plt.annotate(f'{value:.2f}',
                     xy=(df['Métrica'][i], value),
                     xytext=(5, 5),
                     textcoords='offset points')

plt.xlabel('Métricas')
plt.ylabel('Valores')
plt.title('Comparação Completa das Métricas por Tipo de Carga de Trabalho')
plt.legend()
plt.grid(True)
plt.show()
```
